```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny, color="red")
plt.plot(x, cosy, color="black", linestyle="--")
# plt.xlim(-5, 15)
# plt.ylim(0, 1.5)
plt.axis([-1, 11, -2, 2])
plt.show()
```
